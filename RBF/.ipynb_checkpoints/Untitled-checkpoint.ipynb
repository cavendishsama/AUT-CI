# Gaussian Distribution

```python
def gauss(X, mu, sigma):
    return np.exp(-1 / (2 * sigma**2) * (X - mu)**2) / np.sqrt(2 * np.pi * sigma**2)
```

# RBF

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import random

class RBF:
    """Implementation of a Radial Basis Function Network"""
    def __init__(self, x, k=2, lr=0.1, epochs=150):
        self.x = x
        self.k = k
        self.lr = lr
        self.epochs = epochs
        self.kmeans = KMeans(k).fit(np.array(x).reshape(-1, 1))
        
        self.w = np.random.randn(k)
        self.b = np.random.randn(1)
    
    def fit(self, y):
        self.centers = np.array(self.kmeans.cluster_centers_)
        
        ## first approach
#         max_d = max([np.abs(c1 - c2) for c1 in self.centers for c2 in self.centers])
        ## second approach
        max_d = 0
        for i in range(self.k):
            if i == self.k:
                continue
            for c1 in self.centers:
                for c2 in self.centers:
                    d = abs(c1 - c2)
                    if (d > max_d).all():
                        max_d = d

        self.std = np.repeat(max_d / np.sqrt(self.k * 2), self.k)
        
        print(self.std)
        print(self.centers)
#         print(self.centers)
#         print(self.centers.T[0])
        for epoch in range(self.epochs):
            for i in range(self.x.shape[0]):
                # tooye gozaresh kar bood :\
                a = np.array([gauss(self.x[i], c, s) for c,s in zip(self.centers.T[0], self.std)])
                print(a)
                
                F = a.T.dot(self.w) + self.b
#                 print(F)
                cost = (y[i] - F).flatten() ** 2
                print(f'Cost = {cost[0]}')

                #backward pass
                error = -(y[i] - F).flatten()
#                 print(error)
                #online update
                
#                 self.w = np.squeeze(self.w)
#                 print(self.w.shape)
#                 print(f"{self.b} + {self.b.shape}")
#                 print(self.w)
#                 print(self.w.shape)
                self.w = self.w - self.lr * a * error
#                 print(self.w)
#                 print(self.w.shape)
                self.b = self.b - self.lr * error
                print(f"______________________{self.std}")
        
    def predict(self, x):
        y_predicted = []
        for i in range(x.shape[0]):
            a = np.array([gauss(x[i], c, s) for c, s, in zip(self.centers, self.std)])
            F = a.T.dot(self.w) + self.b
            y_predicted.append(F)
#             return np.asarray(y_predicted)
        return y_predicted
```

```python
NUM_SAMPLES = 100
X = np.random.uniform(0., 1., NUM_SAMPLES)
X = np.sort(X, axis=0)
noise = np.random.uniform(-0.1, 0.1, NUM_SAMPLES)
y = np.sin(2 * np.pi * X)  + noise

# k = KMeans(K=2 , max_iters=100, plot=False)
# k.fit(np.asanyarray(list(zip(X,y))))
# k._plot()
rbf = RBF(X, k=2, lr=.01)
rbf.fit(y)
y_pred = rbf.predict(X)

plt.plot(X, y, '-o', label='true')
plt.plot(X, y_pred, '-o', label='RBF-Net')
plt.legend()
plt.tight_layout()
plt.show()
```
